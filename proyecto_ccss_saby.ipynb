```python
#primero definir url
bcrp_url = "https://estadisticas.bcrp.gob.pe/estadisticas/series/mensuales/resultados/PN01770AM/html"
bcrp = pd.read_html(bcrp_url)
bcrp_data = bcrp[1]
bcrp_data
```

```python
bcrp_data.columns
```

b) calcula el promedio móvil 12 meses del índice

```python

# Calcular promedio móvil de 12 meses
bcrp_data["Promedio_Movil_12M"] = bcrp_data["Producto bruto interno y demanda interna (índice 2007=100) - PBI"].rolling(window=12).mean()

# Mostrar resultado
print("Promedio móvil de 12 meses calculado:")
print(bcrp_data[["Producto bruto interno y demanda interna (índice 2007=100) - PBI", "Promedio_Movil_12M"]].tail())
```

c) calcula el promedio móvil 12 meses del índice. A partir de la nueva serie, calcula la variación anual (es decir, por ejemplo, enero de 2019 vs enero de 2018).

```python
bcrp_data["Variacion_Anual"] = bcrp_data["Producto bruto interno y demanda interna (índice 2007=100) - PBI"].pct_change(periods=12) * 100

# Mostrar resultados
print("Promedio móvil de 12 meses y variación anual calculados:")
print(bcrp_data[["Producto bruto interno y demanda interna (índice 2007=100) - PBI", "Promedio_Movil_12M", "Variacion_Anual"]].tail())
```

d) Estima la tasa de crecimiento promedio para el periodo marzo 2015 - febrerp 2020. 

```python
periodo_base = bcrp_data.loc["2015-03":"2020-02"]
tasa_crecimiento_promedio = ((periodo_base["Producto bruto interno y demanda interna (índice 2007=100) - PBI"].iloc[-1] / periodo_base["Producto bruto interno y demanda interna (índice 2007=100) - PBI"].iloc[0]) ** (1 / 5)) - 1
print(f"Tasa de crecimiento promedio anual para el periodo marzo 2015 - febrero 2020: {tasa_crecimiento_promedio:.4f} ({tasa_crecimiento_promedio * 100:.2f}%)")
```

```python
print("Promedio móvil de 12 meses y variación anual calculados:")
print(bcrp_data[["Producto bruto interno y demanda interna (índice 2007=100) - PBI", "Promedio_Movil_12M", "Variacion_Anual"]].tail())
```

e) Luego, utiliza dicho parámetro para calcular el escenario alternativo de la economía peruana. Es decir, la evolución del PBI sin COVID-19. 

```python
#no corre codigo
days_since_base = (bcrp_data.index - pd.Timestamp("2015-03")).days.values / 365.25
bcrp_data["Escenario_Alternativo"] = periodo_base["Producto bruto interno y demanda interna (índice 2007=100) - PBI"].iloc[0] * ((1 + tasa_crecimiento_promedio) ** days_since_base)
```

f) Finalmente, grafiquelos y brinde información sobre la brecha entre los datos reales y la tendencia prepandémica.

```python
 # Graficar el promedio móvil
plt.figure(figsize=(12, 6))
plt.plot(bcrp_data.index, bcrp_data["Producto bruto interno y demanda interna (índice 2007=100) - PBI"], label="PBI Real", color="blue")
plt.plot(bcrp_data.index, bcrp_data["Promedio_Movil_12M"], label="Promedio Móvil 12M", color="orange")

plt.title("PBI y Promedio Móvil de 12 Meses")
plt.xlabel("Fecha")
plt.ylabel("Índice del PBI (2007=100)")
plt.legend()
plt.grid()
plt.show()
```
